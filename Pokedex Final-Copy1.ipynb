```python
#Pokedex
import requests
import json
import os

# Crear la carpeta "pokedex" si no existe
if not os.path.exists("pokedex"):
    os.makedirs("pokedex")

# Obtener el nombre del Pokémon del usuario
nombre_pokemon = input("Introduce el nombre del Pokémon: ").lower()

# Construir la URL de la API
url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"

# Obtener los datos del Pokémon de la API
response = requests.get(url)

# Comprobar si el Pokémon existe
if response.status_code == 200:
    # Obtener los datos del Pokémon
    datos_pokemon = response.json()

    # Obtener la imagen del Pokémon
    imagen_pokemon = datos_pokemon["sprites"]["front_default"]

    # Obtener las estadísticas del Pokémon
    peso_pokemon = datos_pokemon["weight"]
    altura_pokemon = datos_pokemon["height"]
    movimientos_pokemon = [movimiento["move"]["name"] for movimiento in datos_pokemon["moves"]]
    habilidades_pokemon = [habilidad["ability"]["name"] for habilidad in datos_pokemon["abilities"]]
    tipos_pokemon = [tipo["type"]["name"] for tipo in datos_pokemon["types"]]

    # Mostrar la información del Pokémon
    print(f"**Nombre:** {datos_pokemon['name'].capitalize()}")
    print(f"**Imagen:** {imagen_pokemon}")
    print(f"**Peso:** {peso_pokemon} kg")
    print(f"**Altura:** {altura_pokemon} m")
    print("**Movimientos:**")
    for movimiento in movimientos_pokemon:
        print(f" - {movimiento}")
    print("**Habilidades:**")
    for habilidad in habilidades_pokemon:
        print(f" - {habilidad}")
    print("**Tipos:**")
    for tipo in tipos_pokemon:
        print(f" - {tipo}")

    # Guardar la información del Pokémon en un archivo .json
    with open(f"pokedex/{nombre_pokemon}.json", "w") as archivo_json:
        json.dump({"nombre": datos_pokemon["name"], "imagen": imagen_pokemon, "peso": peso_pokemon, "altura": altura_pokemon, "movimientos": movimientos_pokemon, "habilidades": habilidades_pokemon, "tipos": tipos_pokemon}, archivo_json, indent=4)

    print(f"Información del Pokémon guardada en el archivo '{nombre_pokemon}' dentro de la carpeta 'pokedex'.")
else:
    print("El Pokémon no está registrado en la Pokédex.")
```
